```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

# Exploratory analysis

## - Load data

```python
files = (["KingBaseLite2019-A00-A39", "KingBaseLite2019-A40-A79", "KingBaseLite2019-A80-A99",
          "KingBaseLite2019-B00-B19", "KingBaseLite2019-B20-B49", "KingBaseLite2019-B50-B99",
          "KingBaseLite2019-C00-C19", "KingBaseLite2019-C20-C59", "KingBaseLite2019-C60-C99",
          "KingBaseLite2019-D00-D29", "KingBaseLite2019-D30-D69", "KingBaseLite2019-D70-D99",
          "KingBaseLite2019-E00-E19", "KingBaseLite2019-E20-E59", "KingBaseLite2019-E60-E99"])
```

```python
games_plies = []
positions_plies = []
for file_name in files:
    games_plies.append(np.load(file_name + "_games_plies.npy"))
    positions_plies.append(np.load(file_name + "_positions_plies.npy"))
games_plies = np.concatenate(games_plies, axis=0)
positions_plies = np.concatenate(positions_plies, axis=0)
```

## - Number of games, positions and plies

```python
print("Total number of games is " + str(games_plies.shape[0]) + ".")
print("Maximal number of plies in a game is " + str(np.amax(games_plies)) + ".")
print("Total number of extracted positions is " + str(positions_plies.shape[0]) + ".")
print("Total number of extracted examples is " + str(2*positions_plies.shape[0]) + ".")
```

## - Number of plies for games and extracted positions

```python
plt.figure()
plt.hist(games_plies, bins=range(0,302,2), alpha=0.4, density=True, label='Total in game')
plt.hist(positions_plies, bins=range(0,302,2), alpha=0.4, density=True, label='Extracted position')
plt.axis([0,230,0,0.018])
plt.xlabel('Number of plies played')
plt.ylabel('Normed number of games or positions')
plt.title('Density of number of plies played')
plt.legend()
```

```python
plt.figure()
plt.hist(positions_plies, bins=range(0,231), alpha=0.4, density=False, label='Extracted position')
plt.axis([0,230,0,1200000])
plt.xlabel('Number of plies played')
plt.ylabel('Number of positions')
plt.title('Frequency of number of plies played')
plt.legend()
```
